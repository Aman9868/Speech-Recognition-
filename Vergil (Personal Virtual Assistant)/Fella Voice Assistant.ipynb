```python

'''SPEAK by VOICE ASSISTANT'''
def speak(audio):
    engine=p.init() #gives information about current driver class
    voices=engine.getProperty('voices')
    #setter method[0]mal,[1]female
    engine.setProperty('voice',voices[0].id)
    engine.say(audio)  #Method of speaking of the assistant                                                             
    engine.runAndWait()                     

'''Ask Time'''    
def wish():
    hour=int(datetime.now().hour)
    if hour >=0 and hour <12:
        speak('Good Morning Sir')
    elif hour>=12 and hour <18:
            speak('Good Afternoon Sir')
    else:
                speak('Good Evening Sir')
            
            
             

'''Speech Recognizer'''
def takeinput():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        r.pause_threshold = 1
        r.energy_threshold=10000    #Used for hearing low sound
        r.adjust_for_ambient_noise(source,1.2)
        print('Listening...')
        data=""
        audio=r.listen(source)
    try:
        print('Recognizing')
        data=r.recognize_google(audio, language='en-in')
        print(data)
    except sr.UknownValueError:
        print("Sorry I didn't understand,please repeat that again")
    return data
        
def intro():
    speak('My Name is Vergil. How May I Assist You ?')
    
'''Compute INtelligence with wolframaplha api'''
def ask_anything(question):
    try:
        app_id="KUVPHV-Q7A2JWYTGP"
        client=wolframalpha.Client(app_id)
        res=client.query(question)
        answer=next(res.results).text
        print(answer)
        return answer
    except:
        speak('Sorry Sir could not find the results..')
        
    
    
    
'''SENDEMAIL'''
def send_email(receiver,subject,message):
    
    sender='9868638501aman@gmail.com'
    password='Aman9868@'
    receiver="9868638501aman@gmail.com"
    #SMTP CONNECTIONs
    server=smt.SMTP("smtp.gmail.com",587)
    server.ehlo()  #IDENTIFY YOURSELF IN SMTP SERVER
    server.starttls()  # Put the Connection to SMTP server into TLS mode.
    server.ehlo()
    server.login(sender,password)  #Entering  loginCredentials
    
    
    msg=EmailMessage()
    msg['To']=receiver
    msg['From']=sender
    msg['Subject']=subject
    msg.set_content(message)
    
    
    server.send_message(msg)  # Sending Email
    server.close()
```
